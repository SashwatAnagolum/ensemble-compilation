```python
import pennylane as qml
```

```python
qml.data.list_datasets()["qchem"]["BeH2"]
```

```python
dataset = qml.data.load("qchem", molname="BeH2")
```

```python
molecule_data = dataset[0]
```

```python
hamiltonian = qml.qchem.molecular_hamiltonian(molecule_data.molecule, mapping="bravyi_kitaev")
```

```python
terms = []

for term in hamiltonian[0].terms()[1]:
    terms.append(str(term))
```

```python
processed_terms = []

pauli_mapping = {
    "X": 1,
    "Y": 2,
    "Z": 3,
}

for term in terms[1:]:
    qubit_ops = [t.strip() for t in term.split("@")]
    ops_and_targets = [(qubit_op.split("(")[0], qubit_op.split("(")[1][:-1]) for qubit_op in qubit_ops]
    
    processed_term = [0 for _ in range(14)]
    
    for (op, target) in ops_and_targets:
        processed_term[int(target)] = pauli_mapping[op]
        
    processed_terms.append(processed_term)
```

```python
for term in processed_terms:
    print(" " * 4, term, ",", sep="")
```
